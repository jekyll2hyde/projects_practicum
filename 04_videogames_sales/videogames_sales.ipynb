### Изучение наиболее прибыльных платформ за период с 1980 по 2016 гг.

Определим, для каких платформ суммарные продажи игр были наиболее значительными за весь период сбора данных и выберем основные платформы, которые в дальнейшем рассмотрим более подробно:

```python
# Создание функции для построения диаграммы Парето
def build_pareto(data, year):
    # Построение диаграммы Парето по суммарным продажам игр для каждой платформы
    data['cumpercentage'] = data['total_sales'].cumsum()/data['total_sales'].sum()*100
    fig, ax = plt.subplots(figsize = (16,9))
    ax.bar(data.index, data['total_sales'], color="C0")
    ax2 = ax.twinx()
    ax2.plot(data.index, data['cumpercentage'], color="C1", marker="D", ms=7)
    ax2.yaxis.set_major_formatter(PercentFormatter())

    # Настройки основной оси
    ax.tick_params(axis="y", colors="C0")
    ax.set_xlabel('Платформа')
    ax.set_ylabel('Продажи игр, млн.копий')

    # Настройки вспомогательной оси
    ax2.tick_params(axis="y", colors="C1")
    ax2.set_ylabel('Кумулятивная доля от общих продаж')
    plt.title('Суммарные продажи по платформам в {}'.format(year))

    # Вывод диаграммы на экран
    plt.show()
    
# Формирование сводной таблицы с суммарными продажами по платформам
platform_sales = data.pivot_table(
    index='platform',
    values = 'total_sales',
    aggfunc = 'sum').sort_values(
    by = 'total_sales',
    ascending = False)

# Применение функции для построения диаграммы Парето
build_pareto(platform_sales, '1980 - 2016 гг.')
```

Для определения платформ, внесших наибольший вклад в продажи игр за период с 1980 по 2016 год была построена диаграмма Парето, илюстрирующая суммарные продажи для каждой платформы, а также кумулятивную долю от общей суммы продаж по всем платформам.

Из построенной диаграммы видно, что более 60% продаж игр за период с 1980 по 2016 годы приходится на платфомы *PS2, X360, PS3, Wii, DS, PS*. Но, известно, что у геймеров пользуются популярностью и другие платформы, не вошедшие в данный список. Поэтому для дальнейшего, более подробного изучения, выберем платформы, на которые в сумме приходится 80% продаж за рассмотренный период. К ним относятся: *PS2, X360, PS3, Wii, DS, PS, GBA, PS4, PSP, PC, 3DS*.

### Определение срока актуальности платформ

Далее из выбранных платформ отберем те, игры на которые уже не выпускаются, и определим средний срок актуальности платформ после пиковых продаж:

```python
def count_lifecycle(data):
    
    # Создание сводной таблицы ежегодных суммарных продаж по платформам
    annual_sales = data.pivot_table(
        index = ['platform', 'year_of_release'],
        values = 'total_sales',
        aggfunc = 'sum').reset_index()
    
    # Создание сводной таблицы последних лет поддержки платформ
    last_year = data.pivot_table(
        index = 'platform',
        values = 'year_of_release',
        aggfunc = 'max')
    
    # Создание сводной таблицы максимальных суммарных продаж за год по платформам
    max_annual_sales = annual_sales.pivot_table(
        index = 'platform',
        values = 'total_sales',
        aggfunc = 'max').reset_index()
    
    # Добавление в таблицу года пиковых продаж
    max_annual_sales = max_annual_sales.merge(
        annual_sales,
        on=['platform','total_sales'],
        how='left')
    
    # Добавление в таблицу последнего года продаж
    max_annual_sales = max_annual_sales.merge(
        last_year,
        on='platform',
        how='left')

    # Переименование столбцов таблицы
    max_annual_sales.columns = ['platform','max_annual_sales', 'peak_year', 'last_year']
    
    # Добавление в таблицу столбца со сроком поддержки платформ, включающем первый и последний годы
    max_annual_sales['lifecycle'] = max_annual_sales['last_year'] - max_annual_sales['peak_year']+1

    return max_annual_sales

# Применение функции 'count_lifecycle'  
platform_lifecycle = count_lifecycle(data)

# Исключение из таблицы строк с платформами с наименьшими продажами
platform_lifecycle = platform_lifecycle[platform_lifecycle['platform'].isin(best_selling_platforms)]

# Создание функции для присвоения статуса о поддержке платформ
def platform_status(row):
    if row < 2016:
        return 'not supported'
    else:
        return 'supported'
    
# Применение функции    
platform_lifecycle['status'] = platform_lifecycle['last_year'].apply(platform_status)

# Вывод талицы на экран
platform_lifecycle
```

Из полученной таблицы видно, что часть платформ на данный момент уже не поддерживается. Выделим в отдельный датафрейм информацию о неподдерживаемых платформах:

```python
ns_platforms = platform_lifecycle.query('status == "not supported"').reset_index(drop=True)
ns_platforms
```

Если отталкиваться от данных в полученной таблице, то срок актуальности платформ после пиковых продаж варьируется в пределах от 7 до 10 лет.

Рассчитаем средний срок актуальности платформ:

```python
median_lifecycle = ns_platforms['lifecycle'].median()
print('Средний срок актуальности платформы: {} лет'.format(int(median_lifecycle)))
```

### Изучение зависимостей глобальных продаж компьютерных игр от оценок пользователей и критиков

На примере наиболее популярной и потенциально прибыльной платформы, *PS4*, рассмотрим зависимость глобальных продаж игр от оценок игроков и критиков:

```python
def sales_dependance(variable, diag_title):
    actual_data.query('platform == "PS4"').plot(
        kind = 'scatter',
        x = variable,
        y = 'total_sales',
        figsize = (10,6),
        alpha = 0.5
    )
    plt.title(diag_title)
    plt.ylabel('Продажи, млн.копий')
    plt.xlabel('Оценка')
    plt.yticks(np.arange(0, 13, step=1))
    plt.show()
    print('Коэффициент корреляции Пирсона: {:.2f}'.format(actual_data['total_sales'].corr(actual_data[variable])))
    print()
        
sales_dependance('user_score', 'Зависимость продаж игр PS4 от оценки игроков')
sales_dependance('critic_score', 'Зависимость продаж игр PS4 от оценки критиков')
```

Диаграмма зависимости продаж игр от оценки игроков не отражает какой-либо зависимости рассматриваемых величин. Наибольшей плотностью обладает область продаж в пределах 1 млн.копий, остающаяся практически неизменной от оценки 5.5 до 8.8 баллов. Коэффициент корреляции Пирсона в этом случае, равен 0.09, что подтверждает отсутствие зависимости.

На диаграмме зависимости  продаж от оценки критиков наибольшей плотностью обладает область продаж в пределах 1 млн.копий при оценке от 60 до 85 баллов. При этом с оценки 60 баллов наблюдается усиливающийся разброс продаж. Коэффициент корреляции Пирсона в этом случае, равен 0.32, что говорит об умеренной положительной зависимости.

Для того, чтобы представлять зависимости яснее, можно укрупнить данные, использовав при построении диаграмм медианные значения строк столбца *total_sales*, соответствующих уникальным значениям столбцов, зависимость от котрых будет рассчитываться.

```python
# Создание функции для описания зависимости общих продаж игр по платформам от оценок игроков и критиков
def median_sales_dependance(platform):
    # Построение среза по необходимой платформе
    platform_data = actual_data.query('platform == "{}"'.format(platform))
    # Построение сводной таблицы по оценке игроков
    platform_data_pivot1 = platform_data.pivot_table(
        index = 'user_score',
        values = 'total_sales',
        aggfunc = 'median').reset_index()
    # Построение диаграммы рассеяния
    platform_data_pivot1.plot(
        kind = 'scatter',
        x = 'user_score',
        y = 'total_sales',
        figsize=(10,6)
    )
    plt.title('Зависимость средних продаж игр {} от оценки игроков'.format(platform))
    plt.ylabel('Продажи, млн.копий')
    plt.xlabel('Оценка')
    plt.yticks(np.arange(0, 13, step=1))
    plt.show()
    # Вывод на экран коэффициента корреляции
    print('Коэффициент корреляции Пирсона: {:.2f}'.format(
        platform_data_pivot1['total_sales'].corr(
            platform_data_pivot1['user_score']))
         )
    print()
    # Построение сводной таблицы по оценке критиков
    platform_data_pivot2 = platform_data.pivot_table(
        index = 'critic_score',
        values = 'total_sales',
        aggfunc = 'median').reset_index()
    # Построение диаграммы рассеяния
    platform_data_pivot2.plot(
        kind = 'scatter',
        x = 'critic_score',
        y = 'total_sales',
        figsize=(10,6)
    )
    plt.title('Зависимость средних продаж игр {} от оценки критиков'.format(platform))
    plt.ylabel('Продажи, млн.копий')
    plt.xlabel('Оценка')
    plt.yticks(np.arange(0, 13, step=1))
    plt.show()
    # Вывод на экран коэффициента корреляции
    print('Коэффициент корреляции Пирсона: {:.2f}'.format(
        platform_data_pivot2['total_sales'].corr(
            platform_data_pivot2['critic_score']))
         )
    print()
# Применение функции    
median_sales_dependance('PS4')
```

Укрупнение данных подтвердило отсутствие зависимости между оценкой игроков и продажами игр, а также позволило получить более выраженную картину и более высокий коэффициент корреляции Пирсона для зависимости продаж от оценки критиков.

Из диагарммы зависимости продаж от оценки критиков видно, что при оценке игры до 75 баллов не наблюдается роста в продажах игр, а с ростом оценки выше 75 баллов количество продаж увеличивается.

Можно подробнее рассмотреть участок оценок от 75 баллов и сответствующие данным оценкам медианы общих продаж:

```python
# Создание функции для описание зависмости общих продаж иг разных платформ от оценки критиков свыше 75 баллов
def dependance_on_highmarks(platform):
    # Построение среза данных по необходимой платформе и оценке критиков свыше 75 баллов
    platform_highrated = actual_data.query('critic_score > 75 and platform == "{}"'.format(platform))
    pivot = platform_highrated.pivot_table(
        index = 'critic_score',
        values = 'total_sales',
        aggfunc = 'median'
    ).reset_index()
    # Построение диаграммы рассеяния
    pivot.plot(
    kind = 'scatter',
    x = 'critic_score',
    y = 'total_sales',
    figsize = (10,6))
    plt.title('Зависимость средних продаж игр {} от оценки критиков свыше 75 баллов'.format(platform))
    plt.ylabel('Продажи, млн.копий')
    plt.xlabel('Оценка')
    plt.yticks(np.arange(0, 13, step=1))
    plt.show()
    # Вывод коэффициента корелляции
    print('Коэффициент корреляции Пирсона: {:.2f}'.format(
        pivot['total_sales'].corr(
            pivot['critic_score']))
         )
          
dependance_on_highmarks('PS4')
```

Таким образом при оценке критиков от 75 и выше наблюдается сильная положительная связь между оценкой критиков и продажами игр, описываемая коэффициентом корреляции Пирсона = 0.72.

Проверим, как оценки и гроков и критиков влияют на общие продажи игр для других потенциально прибыльных платформ:

```python
median_sales_dependance('XOne')
median_sales_dependance('3DS')
median_sales_dependance('PC')
```

В сравнении с выводами, полученными при рассмотрении зависимостей между оценками игроков и критиков на продажи игр для *PS4*, зависимости, полученные при рассмотрении показателей других потенциально прибыльных платформ, ведут себя аналогично:

 - Связь продаж с оценкой игроков слабая;
 - Заментная связь продаж с оценкой критиков наблюдается при оценке свыше 75 баллов.
 
Исключение составляет платформа *3DS*, где связь продаж игр с оценкой игроков описывается не слабой, а умеренной зависимостью.
 
Рассмотрим более подробно для каждой потенциально прибыльной платформы участки оценок от 75 баллов и сответствующие данным оценкам медианы общих продаж:

```python
dependance_on_highmarks('XOne')
dependance_on_highmarks('3DS')
dependance_on_highmarks('PC')
```

В итоге, при оценке критиков от 75 и выше наблюдается заметная положительная связь между оценкой и продажами игр, описываемая коэффициентами корреляции Пирсона от 0.54 до 0.65.

###### Вывод

У самой популярной и потенциально прибыльной платфомы, *PS4*, продажи игр не связаны с оценками игроков, а между оценкой критиков и продажами сильная связь (К = 0.72) наблюдается от оценки 75 и выше.

### Изучение влияния рейтинга ESRB на продажи в регионах

ESRB - негосударственная организация, основное направление деятельности которой - принятие и определение рейтингов для компьютерных видеоигр и другого развлекательного программного обеспечения в Северной Америке.

Опишем обозначения рейтингов ESRB:
 - *EC («Early childhood»)* — *«Для детей младшего возраста»*: игра подходит для детей от 3 лет и не содержит материалов, которые родители могли бы счесть неподходящими. Обычно представляют собой развивающие игры;
 - *E («Everyone»)* — *«Для всех»*: содержание вполне подходит для возрастной категории от 6 лет. Такие игры могут понравиться и взрослым. Игры с этим рейтингом могут содержать минимальное насилие, в основном «мультяшного» характера;
 - *E10+» («Everyone 10 and older»)* — *«Для всех от 10 лет и старше»*: проекты с данным рейтингом могут содержать или немного больше мультипликационного или мягкого насилия, или несколько откровенные сцены, или минимальное количество крови;
 - *T («Teen»)* — *«Подросткам»*: игра подходит для лиц от 13 лет. Проекты из данной категории могут содержать насилие, непристойные сцены, грубый юмор, в меру откровенное сексуальное содержимое, кровь или нечастое использование ненормативной лексики;
 - *M («Mature»)* — *«Для взрослых»*: материалы игры не подходят для подростков младше 17 лет. Проекты с данным рейтингом могут содержать достаточно жестокое насилие, большое количество крови с расчленением, непристойные сексуальные сцены или грубую ненормативную лексику, нежелательную для младшей аудитории;
 - *RP («Rating Pending»)* — *«Рейтинг ожидается»*: продукт был отправлен в ESRB и ожидает присвоения рейтинга. Обозначение используется только на рекламных презентациях и в демо-версиях игр до официальной даты выпуска в продажу.

Рассмотрим влияние рейтинга ESRB на продажи в отдельных регионах:

```python
# Создание функции для вывода на экран продаж игр с определенным рейтингом ESRP
def esrp_data(region_sales):
    # Заполнение пропущенных значений столбца 'rating' значением 'RP'
    actual_data['rating'] = actual_data['rating'].fillna('unknown')
    # Вывод сводной таблицы на экран
    display(
        actual_data.pivot_table(
            index = 'rating',
            values = region_sales,
            aggfunc = 'sum').sort_values(
            by=region_sales,
            ascending = False))
    
# Применение функции    
esrp_data('na_sales')
esrp_data('eu_sales')   
esrp_data('jp_sales') 
esrp_data('total_sales') 
```

Возможно, большая часть игр с неизвестным рейтингом - это японские игры, предназначенные только для японского рынка и имеющие на японском рынке наибольшую популярность. Посмотрим, какая доля игр c неизвестным рейтингом продается только в Японии:

## Проверка гипотез

Одной из поставленных задач в данном проекте является проверка следующих гипотез:

1. Средние пользовательские рейтинги платформ *XOne* и *PC* одинаковые;
2. Средние пользовательские рейтинги жанров Action и Sports разные.

Сравним средние пользовательские рейтинги платформ и оценим статистическую значимость полученной разницы:

```python
# Расчет и вывод на экран среднего пользовательского рейтинга по платформам с помощью сводной таблицы:
actual_data.pivot_table(
    index = 'platform',
    values = 'user_score',
    aggfunc = 'mean').reset_index().query(
    'platform == ["XOne", "PC"]'
)
```

Из результатов расчета видно, что в рамках рассмотренной выборки средний пользовательский рейтинг платформы *PC* больше, чем *XOne* на 0.25. Проверим, значимо ли статистически данное различие.

Проверка гипотезы о равенстве средних пользовательских рейтингов для каждой из платформ будет сводиться к проверке гипотезы о равенстве средних значений двух генеральных совокупностей по взятым из них выборкам.

Таким оборазом, из полученных нами ранее данных нужно сформировать отдельные выборки по пользовательским рейтингам каждой из платформ:

```python
# Формирование выборки, содержащей данные о пользовательском рейтинге платформы 'PC'
score_data_pc = actual_data[
    (actual_data['platform'] == 'PC') & 
    (actual_data['user_score'].notnull())]['user_score']
print(score_data_pc.describe())
print()

# Формирование выборки, содержащей данные о пользовательском рейтинге платформы 'XOne'
score_data_xone = actual_data[
    (actual_data['platform'] == 'XOne') &
    (actual_data['user_score'].notnull())]['user_score']
print(score_data_xone.describe())
```

Каждая из полученных выборок характеризует свою генеральную совокупность.

Сформулируем нулевую гипотезу, относительно которой мы будем проверять статистическую значимость результатов сравнения пользовательских рейтингов по платформам:

"*Средние значения двух генеральных совокупностей, характеризуемых выборками ***score_data_pc*** и ***score_data_xone***,  равны"*.

Тогда альтернативной гипотезой будет:

"*Средние значения двух генеральных совокупностей, характеризуемых выборками ***score_data_pc*** и ***score_data_xone***,  не равны"*.

Для проверки гипотезы применим *t-test*. Критерий значимости, *alpha*, примем равным 0.05.

```python
# Создадим функцию для тестирования

def t_test(sample_1, sample_2, var_equality):

    # Установка критерия значимости
    alpha = 0.05

    # Применение теста
    results = st.ttest_ind(
    sample_1,
    sample_2,
    equal_var = var_equality)

    # Вывод на экран результата теста
    print('p-значение: ', results.pvalue)
    if results.pvalue < alpha:
        print('Отвергаем нулевую гипотезу')
    else:
        print('Не отвергаем нулевую гипотезу')
        
# Применение функции
t_test(score_data_pc, score_data_xone, True)
# equal_var - устанавливается True, т.к. выборки взяты из схожих по параметрам совокупностей
```

Из результатов теста можно сделать вывод о том, что вероятность получить наблюдаемый результат при верной нулевой гипотезе выше критерия значимости, и мы не можем отвергнуть нулевую гипотезу, что позволяет нам считать, что средние пользовательские рейтинги платформ PC и XOne не отличаются.

Проверим гипотезу об отличии средних пользовательских рейтингов жанров *Action* и *Sports*. Сравним средние пользовательские рейтинги жанров и оценим статистическую значимость полученной разницы:

Каждая из полученных выборок характеризует свою генеральную совокупность.

Сформулируем нулевую гипотезу, относительно которой мы будем проверять статистическую значимость результатов сравнения пользовательских рейтингов по платформам:

"*Средние значения двух генеральных совокупностей, характеризуемых выборками ***score_data_action*** и ***score_data_sports***,  равны"*.

Тогда альтернативной гипотезой будет:

"*Средние значения двух генеральных совокупностей, характеризуемых выборками ***score_data_action*** и ***score_data_sports***,  не равны"*.

Во всех трех регионах наименее популярными или непродающимися вообще являются игры, ожидающие присвоения рейтинга, *RP*, и игры с рейтингом *EС («Early childhood», «Для детей младшего возраста»)*, подходящие для детей от 3 лет и представляющие, в основном, развивающие игры.

 2. **Оценка критиков:** *Максимальная*;
 3. **Жанр:** *Action*;
